## Linear regression

```python
def standard_units(any_numbers):
    "Convert any array of numbers to standard units."
    return (any_numbers - np.mean(any_numbers)) / np.std(any_numbers)  

def correlation(t, x, y):
    """Return the correlation coefficient (r) of two variables."""
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))

def slope(t, x, y):
    """The slope of the regression line (original units)."""
    r = correlation(t, x, y)
    return r * np.std(t.column(y)) / np.std(t.column(x))

def intercept(t, x, y):
    """The intercept of the regression line (original units)."""
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))
```

## Least squares

```python
little_women = Table.read_table('http://inferentialthinking.com/notebooks/little_women.csv')
little_women = little_women.move_to_start('Periods')
little_women.show(3)
```

```python
little_women.scatter('Periods', 'Characters')
```

```python
correlation(little_women, 'Periods', 'Characters')
```

```python
def linear_predict(table, x, y):
    """Return the height of the regression line at each x value."""
    a = slope(table, x, y)
    b = intercept(table, x, y)
    return a * table.column(x) + b
```

```python
lw_predicted = little_women.with_column(
    'Predicted', linear_predict(little_women, 'Periods', 'Characters')
)
lw_predicted.scatter('Periods')
```

```python
sample = [[131, 14431], [231, 20558], [392, 40935], [157, 23524]]
def lw_errors(slope, intercept):
    print('Slope:    ', np.round(slope), 'characters per period')
    print('Intercept:', np.round(intercept), 'characters')
    little_women.scatter('Periods', 'Characters')
    xlims = np.array([50, 450])
    plots.plot(xlims, slope * xlims + intercept, lw=2)
    for x, y in sample:
        plots.plot([x, x], [y, slope * x + intercept], color='r', lw=2)
```

```python
lw_reg_slope = slope(little_women, 'Periods', 'Characters')
lw_reg_intercept = intercept(little_women, 'Periods', 'Characters')
lw_errors(lw_reg_slope, lw_reg_intercept)
```

```python
lw_errors(50, 10000)
```

```python
lw_errors(-100, 50000)
```

```python
def lw_rmse(slope, intercept):
    lw_errors(slope, intercept)
    x = little_women.column('Periods')
    y = little_women.column('Characters')
    predicted = slope * x + intercept
    mse = np.mean((y - predicted) ** 2)
    print("Root mean squared error:", mse ** 0.5)
```

```python
lw_rmse(50, 10000)
```

```python
lw_rmse(-100, 50000)
```

```python
lw_rmse(90, 4000)
```

```python
lw_rmse(lw_reg_slope, lw_reg_intercept)
```

## Non-linear regression

```python
shotput = Table.read_table('http://inferentialthinking.com/notebooks/shotput.csv')
```

```python
shotput.scatter('Weight Lifted')
```

```python
slope(shotput, 'Weight Lifted', 'Shot Put Distance')
```

```python
intercept(shotput, 'Weight Lifted', 'Shot Put Distance')
```

```python
def shotput_linear_mse(any_slope, any_intercept):
    x = shotput.column('Weight Lifted')
    y = shotput.column('Shot Put Distance')
    predicted = any_slope*x + any_intercept
    return np.mean((y - predicted) ** 2)
```

```python
minimize(shotput_linear_mse)
```

```python
shotput.with_column(
    'Best Straight Line', linear_predict(shotput, 'Weight Lifted', 'Shot Put Distance')
).scatter('Weight Lifted')
```

```python
def shotput_quadratic_mse(a, b, c):
    x = shotput.column('Weight Lifted')
    y = shotput.column('Shot Put Distance')
    predicted = a*(x**2) + b*x + c
    return np.mean((y - predicted) ** 2)
```

```python
best = minimize(shotput_quadratic_mse)
best
```

```python
x = shotput.column('Weight Lifted')
quadratic_curve = best.item(0)*(x**2) + best.item(1)*x + best.item(2)
```

```python
shotput.with_column('Best Quadratic Curve', quadratic_curve).scatter('Weight Lifted')
```
